```python
import os
import numpy as np
import shutil

path= '../data/raw/IDC_regular_ps50_idx5'
patients= os.listdir(path)

patient = 10305 #np.random.choice(patients)
print(patient)

def get_patient_coords(patient):
    path0 = os.path.join(path, str(patient))
    patches = os.listdir(path0)

    def find_coord(patches):
        coords = []

        for patch in patches:
            patch_splitted = patch.split("_")
            x = (patch_splitted[2])[1:]
            y = (patch_splitted[3])[1:]
            
            coords.append((patch, (int(x),int(y))))
        return coords

    coords = find_coord(patches)

    return coords
```

```python
from PIL import Image, ImageDraw

def show_patient_image(patient):
    images = get_patient_coords(patient)
    print(images)

        # Get max width and height for canvas
    max_x = max(x for _, (x, _) in images)
    max_y = max(y for _, (_, y) in images)

    image_path = os.path.join(path, str(patient) ,images[0][0])
    img_sample = Image.open(image_path)
    img_w, img_h = img_sample.size

    canvas = Image.new("RGB", (max_x + img_w, max_y + img_h), "white")

    # Paste images at coordinates
    for image, (x, y) in images:
        image_path = os.path.join(path, str(patient), image)
        img = Image.open(image_path)
        canvas.paste(img, (x, y))
        

    return canvas

```

```python
f = open("../data/deployement/test.txt")
texte = f.read()
test_set = texte.split("\n")[:-1]

for patient_id in test_set:
    patient_id = int(patient_id)
    img = show_patient_image(patient_id)
    img.save(f'../data/deployement/full_images/img_{patient_id}.png')
```

```python
def crop_slice_image(image, id):
    images=[]
                                        
    img_sample = image.convert("RGB")
    img_w, img_h = img_sample.size

    img_sample = img_sample.crop((1,1,img_w , img_h ))
    
    i=0
    for x in range(0,img_w,50):
        for y in range(0,img_h,50):
            patch = img_sample.crop((x,y,x+50,y+50))
            patch = patch.convert('RGB')

            name = f'{id}_x{x}_y{y}'

            patch = np.array(patch)
            if not np.all(patch==255) and not np.all(patch==0):
                images.append((patch,name))

                os.makedirs('../data/testings', exist_ok=True)
                Image.fromarray(patch).save(f'../data/testings/{name}.png')
                i+=1
    print(i)
    return images


img_test = Image.open("../data/deployement/full_images/img_8918.png")
crop_slice_image(img_test, 8918)
```

```python
import pickle

path = '../data/clean/test'
file = open("../data/deployement/test.txt")
texte = file.read()
test_set = set(int(i) for i in texte.split("\n")[:-1])
print(test_set)

dict_of_matrices = {}

def make_matrice(coords):
    
    xmax = np.max([x for x,_,_ in coords])
    ymax = np.max([y for _,y,_ in coords])
    A = np.ones((ymax,xmax))*2
    for x,y,i in coords:
        if i==0:
            A[y-1][x-1]=0
        else:
            A[y-1][x-1]=1
    return A


for id in test_set:
    path_to_images = os.path.join('../data/raw/IDC_regular_ps50_idx5',f'{id}')
    coords = []
    for img in os.listdir(path_to_images):
        x = int((img.split('_')[2])[1:]) - 1
        y = int((img.split('_')[3])[1:]) - 1
        state = (img.split('_')[4])[5]
        coords.append((x//50,y//50,state))
    mat = make_matrice(coords)
    dict_of_matrices[id] = mat

with open("matrices.pkl", "wb") as f:
    pickle.dump(dict_of_matrices, f)
    
    
```
